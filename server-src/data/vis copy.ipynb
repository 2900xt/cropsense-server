```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_plant_001.csv')
df = df[1000:]
df = df[::100]

df2 = pd.read_csv('data_plant_002.csv')
df2 = df2[1000:]
df2 = df2[::100]
```

```python
df.head()
```

```python
df_without_time = df.drop(columns=['Timestamp', 'Plant ID', 'Disease Status'])
df_without_time.corr()

# draw fancy sns heatmap
import seaborn as sns
plt.figure(figsize=(12,10))
sns.heatmap(df_without_time.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix for Plant 001')
plt.show()
```

```python
#graphs of temperature vs. time, humidity vs. time, pressure vs. time, and gas resistance vs. time
#have no x axis markers
plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
plt.plot(df['Timestamp'], df['Temperature (°C)'])
plt.title('Temperature vs Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')

plt.subplot(2, 2, 2)
plt.plot(df['Timestamp'], df['Humidity (%)'])
plt.title('Humidity vs Time')
plt.xlabel('Time')
plt.ylabel('Humidity (%)')

plt.subplot(2, 2, 3)
plt.plot(df['Timestamp'], df['Pressure (kPa)'])
plt.title('Pressure vs Time')
plt.xlabel('Time')
plt.ylabel('Pressure (kPa)')

plt.subplot(2, 2, 4)
plt.plot(df['Timestamp'], df['Gas Resistance (KOhm)'])
plt.title('Gas Resistance vs Time')
plt.xlabel('Time')
plt.ylabel('Gas Resistance (KOhm)')

plt.tight_layout()
plt.show()
```
